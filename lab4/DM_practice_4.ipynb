#### **Система Лоренца:**

Система Лоренца описується трьома диференціальними рівняннями:

$$
\begin{cases}
\frac{dx}{dt} = \sigma (y - x) \\
\frac{dy}{dt} = x (\rho - z) - y \\
\frac{dz}{dt} = xy - \beta z
\end{cases}
$$

де:
- $( \sigma )$ — коефіцієнт Прандтля (зазвичай $ \sigma = 10 $),
- $( \rho )$ — числовий параметр для турбулентності (зазвичай $ \rho = 28 $),
- $( \beta )$ — коефіцієнт відношення висоти до довжини (зазвичай $ \beta = 8/3 $).

#### **Об'єкти Нарендра-Пархтізаратхі:**

Об'єкти Нарендра-Пархтізаратхі описуються наступними рівняннями:

$$
\begin{cases}
x_1(t+1) = \alpha_1 x_1(t) + \beta_1 x_2(t) + \gamma_1 x_3(t) + \delta_1 \\
x_2(t+1) = \alpha_2 x_1(t) + \beta_2 x_2(t) + \gamma_2 x_3(t) + \delta_2 \\
x_3(t+1) = \alpha_3 x_1(t) + \beta_3 x_2(t) + \gamma_3 x_3(t) + \delta_3
\end{cases}
$$

де $(\alpha_i)$, $(\beta_i)$, $(\gamma_i)$ і $(\delta_i)$ — коефіцієнти для кожного з рівнянь.

Їх поведінка моделюється нелінійними диференціальними рівняннями, що описують динаміку в часі. Ці рівняння створюють цікаву, нелінійну та хаотичну поведінку системи.

### **Нелінійних рівнянь Нарендра-Партхізаратхі:**

$$

\dot{x}_1 = a_1 \sin(x_2) - b_1 x_1^2

$$
\dot{x}_2 = a_2 x_1 x_2 - b_2 \cos(x_1)

6. $$

\dot{x}_2 = a_2 \sin(x_1) - b_2 x_2^2

7. $$
\dot{x}_1 = -a_1 x_1^2 + b_1 \cos(x_2)
$$
$$
\dot{x}_2 = a_2 x_1 - b_2 \sin(x_2)
$$

8. $$
\dot{x}_1 = a_1 \sin(x_2) - b_1 x_1^2

\dot{x}_2 = a_2 \cos(x_1) - b_2 \sin(x_2)

### Параметри:
- $( a_1, a_2 )$, $( b_1, b_2 )$ — це коефіцієнти, які можуть бути варійовані для дослідження динаміки системи та впливу на прогнозування.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from scipy.integrate import solve_ivp
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error,  r2_score
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
```

```python
sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0


def lorenz(t, state):
    x, y, z = state
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

initial_state = [1.0, 1.0, 1.0]


t_start = 0.0
t_end = 100.0
n_points = 10000  
t_eval = np.linspace(t_start, t_end, n_points)

sol = solve_ivp(lorenz, [t_start, t_end], initial_state, t_eval=t_eval)

lorenz_data = np.column_stack((sol.y[0], sol.y[1], sol.y[2]))

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(lorenz_data[:, 0], lorenz_data[:, 1], lorenz_data[:, 2])
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.title("Lorentz system")
plt.show()
```

```python
alpha = np.random.uniform(-1, 1, 3)
beta = np.random.uniform(-1, 1, 3)
gamma = np.random.uniform(-1, 1, 3)
delta = np.random.uniform(-0.5, 0.5, 3)

initial_state = [1.0, 1.0, 1.0]

n_points = 10000

narendra_parthasarathy_data = np.zeros((n_points, 3))

narendra_parthasarathy_data[0] = initial_state

for t in range(1, n_points):
    x1_t, x2_t, x3_t = narendra_parthasarathy_data[t-1]
    x1_t1 = alpha[0] * x1_t + beta[0] * x2_t + gamma[0] * x3_t + delta[0]
    x2_t1 = alpha[1] * x1_t + beta[1] * x2_t + gamma[1] * x3_t + delta[1]
    x3_t1 = alpha[2] * x1_t + beta[2] * x2_t + gamma[2] * x3_t + delta[2]
    narendra_parthasarathy_data[t] = [x1_t1, x2_t1, x3_t1]

plt.plot(narendra_parthasarathy_data[:, 0], label="x1")
plt.plot(narendra_parthasarathy_data[:, 1], label="x2")
plt.plot(narendra_parthasarathy_data[:, 2], label="x3")
plt.xlabel("Time step")
plt.ylabel("State values")
plt.title("Objects by Narendra-Parkhtizarathi")
plt.legend()
plt.show()

```

```python
train_val_data, test_data = train_test_split(lorenz_data, test_size=0.15, random_state=42)

train_data, val_data = train_test_split(train_val_data, test_size=0.1765, random_state=42)


print(f"Train data size: {train_data.shape}")
print(f"Validation data size: {val_data.shape}")
print(f"Test data size: {test_data.shape}")

scaler = StandardScaler()

train_data_scaled = scaler.fit_transform(train_data)

val_data_scaled = scaler.transform(val_data)
test_data_scaled = scaler.transform(test_data)

print("mean value after normalization (Train):", np.mean(train_data_scaled, axis=0))
print("mean value after normalization (Validation):", np.mean(val_data_scaled, axis=0))
print("mean value after normalization (Test):", np.mean(test_data_scaled, axis=0))

print("Standard deviation after normalization (Train):", np.std(train_data_scaled, axis=0))
```

```python
linear_model = LinearRegression()
linear_model.fit(train_data_scaled, train_data_scaled[:, 0]) 
train_pred_linear = linear_model.predict(train_data_scaled)
print(f"Linear Regression MSE (Train): {mean_squared_error(train_data_scaled[:, 0], train_pred_linear)}")


tree_model = DecisionTreeRegressor(random_state=42)
tree_model.fit(train_data_scaled, train_data_scaled[:, 0])
train_pred_tree = tree_model.predict(train_data_scaled)
print(f"Decision Tree MSE (Train): {mean_squared_error(train_data_scaled[:, 0], train_pred_tree)}")

random_forest_model = RandomForestRegressor(n_estimators=100, random_state=42)
random_forest_model.fit(train_data_scaled, train_data_scaled[:, 0])
train_pred_forest = random_forest_model.predict(train_data_scaled)
print(f"Random Forest MSE (Train): {mean_squared_error(train_data_scaled[:, 0], train_pred_forest)}")


svm_model = SVR(kernel='rbf')
svm_model.fit(train_data_scaled, train_data_scaled[:, 0])
train_pred_svm = svm_model.predict(train_data_scaled)
print(f"SVM MSE (Train): {mean_squared_error(train_data_scaled[:, 0], train_pred_svm)}")
```

```python


ridge_model = Ridge()
ridge_params = {'alpha': [0.1, 1.0, 10.0]} 
grid_ridge = GridSearchCV(ridge_model, ridge_params, cv=5, scoring='neg_mean_squared_error')
grid_ridge.fit(train_data_scaled, train_data_scaled[:, 0])
print(f"Best params for Ridge: {grid_ridge.best_params_}, Best MSE: {-grid_ridge.best_score_}")

tree_model = DecisionTreeRegressor(random_state=42)
tree_params = {
    'max_depth': [5, 10, 15, None],
    'min_samples_split': [2, 10, 20],
    'min_samples_leaf': [1, 5, 10]
}
grid_tree = GridSearchCV(tree_model, tree_params, cv=5, scoring='neg_mean_squared_error')
grid_tree.fit(train_data_scaled, train_data_scaled[:, 0])
print(f"Best params for Decision Tree: {grid_tree.best_params_}, Best MSE: {-grid_tree.best_score_}")

forest_model = RandomForestRegressor(random_state=42)
forest_params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20, None],
    'min_samples_split': [2, 10, 20],
    'min_samples_leaf': [1, 5, 10]
}
grid_forest = GridSearchCV(forest_model, forest_params, cv=5, scoring='neg_mean_squared_error')
grid_forest.fit(train_data_scaled, train_data_scaled[:, 0])
print(f"Best params for Random Forest: {grid_forest.best_params_}, Best MSE: {-grid_forest.best_score_}")

svm_model = SVR()
svm_params = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}
grid_svm = GridSearchCV(svm_model, svm_params, cv=5, scoring='neg_mean_squared_error')
grid_svm.fit(train_data_scaled, train_data_scaled[:, 0])
print(f"Best params for SVM: {grid_svm.best_params_}, Best MSE: {-grid_svm.best_score_}")
```

### **7. Висновки**

- Найкращою моделлю є **Random Forest**, який продемонстрував найнижчу середню квадратичну помилку на тестовому наборі \( MSE = 4.67e-06 \) та високий коефіцієнт детермінації \( R² = 0.999995 \).
  
- **Ridge Regression** також показала дуже близькі до ідеальних результати, хоча можуть виникати проблеми з переобладнанням через високу точність на навчальних даних.

- **Decision Tree** також добре працювало, хоча його точність була трохи нижчою, ніж у Випадкового лісу.


```python
a1 = 1.0
a2 = 1.0
b1 = 1.0
b2 = 1.0

def derivatives(state, t):
    x1, x2 = state
    dx1_dt = a1 * np.sin(x2) - b1 * x1**2
    dx2_dt = -a2 * x2 + b2 * x1 * np.cos(x2)
    return np.array([dx1_dt, dx2_dt])

t = np.linspace(0, 50, 1000)
initial_state = [1.0, 1.0]

from scipy.integrate import odeint

solution = odeint(derivatives, initial_state, t)

plt.figure(figsize=(10, 5))
plt.plot(t, solution[:, 0], label='x1(t)')
plt.plot(t, solution[:, 1], label='x2(t)')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Dynamics of the Narendra-Parthisarathi system')
plt.legend()
plt.grid()
plt.show()
```
